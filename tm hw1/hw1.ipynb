```python
import numpy as np
import pandas as pd
```

зададим переменные, которые всегда будут константо

```python
Vx=3
Ax=0
Ay=8

```

Начнем цикл, который будет высчитывать все значения для t=[-5..5]

```python
t = -5
while t!=6: 
    print('Расчеты для t=', t)
    Vy=8*t 
    Vy= float('{:.4f}'.format(Vy))
    V=np.sqrt(Vx**2 + Vy**2)
    V= float('{:.4f}'.format(V))
    "найдем моудель ускорения точки"
    A = np.sqrt(Ax**2 + Ay**2)
    A= float('{:.4f}'.format(A))
    "найдем модуль касательного ускорения точки"
    Aτ= abs((Vx*Ax+Vy*Ay)/V)
    Aτ= float('{:.4f}'.format(Aτ))
    "найдем модуль нормального ускорения точки"
    An=np.sqrt(A**2 - Aτ**2)
    An= float('{:.4f}'.format(An))
    "найдем osculating circle"
    p=(V**2 / An)
    p= float('{:.4f}'.format(p))
    k=1/p
    k= float('{:.4f}'.format(k))
    "Выведим значения"
    print(' V = ', V ,'\n', 'a = ', A,'\n', 'aτ= ', Aτ, '\n', 'an= ', An, '\n', 'k = ', k)
    t= t+0.5
```
