```python
import matplotlib.pyplot as plt
import math
ma = 1
mb = 3
md = 20 
Rb = 20
Rd = 20 
iBx = 18
phi = 0.6
g = 9.8
rb = 16 
FK = 64
h = 1
va = []
s = []
for i in range(0, 5): 
    sn = h-i
    s.append(sn)
    van=(ma*sn + md*g*(FK+Rb-math.sqrt(FK**2 - Rb**2)))/(0.5*ma+0.5*mb*iBx/rb**2 + 0.5*md*Rb**2/rb**2 + 0.25*md*Rb**2/((Rd**2)*(rb**2)))
    va.append(math.sqrt(van))
plt.plot(s, va)
plt.xlabel('Va')
plt.ylabel('S')
plt.legend()
plt.show()
```
